### Questão 01

Qual estrutura de dados pode ser colocada na variável "estrutura" do código a seguir para criar uma Series corretamente?

    ss = pd.Series(estrutura)

```python
import pandas as pd
import numpy as np

estrutura = [1, 2, 3]
ss = pd.Series(estrutura)
print(ss)

estrutura2 = np.array(estrutura)
ss2 = pd.Series(estrutura2)
print(ss2)

estrutura3 = {0:'a', '1': 'b', 3: 'c'}
ss3 = pd.Series(estrutura3)
print(ss3)
```

### Questão 02

Qual alternativa contém uma sintaxe INCORRETA para acessar determinada parte de um DataFrame?

```python
dict_medidas = {'peso': [72, 80, 60, 90, 100],
                'altura': [180, 170, 175, 174, 185],
                'idade': [31, 21, 16, 28, 23]}
df = pd.DataFrame(dict_medidas)

print(df)

print('-'*32)
linha = df.loc[0]
print('linha 0\n', linha)

print('-'*32)
col = df['peso']
print('coluna peso\n', col)

print('-'*32)
elemento = df.loc[0, 'peso']
print('linha 0, coluna peso\n', elemento)

print('-'*32)
elemento = df[0, 'peso']
print('linha 0, coluna peso\n', elemento)
```
