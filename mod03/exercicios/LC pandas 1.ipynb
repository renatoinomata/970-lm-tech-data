## Questão 1
Em um concurso público foram contabilizados os números de pessoas inscritas (até a data inicialmente estabelecida para a inscrição) segundo os níveis de escolaridade: fundamental completo, médio completo, superior completo e pós-graduação completa. Segue abaixo a tabela com os valores observados.

| Nível de escolaridade  | Inscritos |
|------------------------|-----------|
| Fundamental completo   | 451 |
| Médio completo         |627 |
| Superior completo      | 292 |
|Pós-graduação completa  | 95 |

Com base na tabela acima, crie uma Series com esses valores e calcule:

* a) O número total de inscritos

* b) Número inscritos que possuem, pelo menos, o superior completo.

* c) Suponha que a inscrição foi prorrogada e, com isso, foi obtido um número adicional de inscrições: 68, 93, 22 e 0 inscritos a mais em cada uma das categorias de escolaridade apresentados na tabela acima (na mesma ordem). Assim, crie uma nova Series com esses valores.

* d) Utilizando as duas Series que você tem, calcule o número total de inscritos após a prorrogação do período de inscrições.


```python
import pandas as pd

inscritos = pd.Series({'fund_comp': 451,
'med_comp': 627,
'sup_comp': 292,
'pos_comp': 95})

print('Inscritos:')
print(inscritos)
print('-'*32)

# a) Número total de inscritos
print('a) Número total de inscritos:', inscritos.sum())
print('-'*32)

# b) Número total de inscritos que possuem, pelo menos, o superior completo
print('b) Número total de inscritos que possuem, pelo menos, o superior completo:',
inscritos['sup_comp':].sum())
print('-'*32)

# c) Suponha que a inscrição foi prorrogada e, com isso, foi obtido um número adicional de inscrições: 68, 93, 22 e 0 inscritos a mais em cada uma das categorias de escolaridade apresentados na tabela acima (na mesma ordem). Assim, crie uma nova Series com esses valores.
print('c) Novos inscritos:')
novos_inscritos = pd.Series({'fund_comp': 68,
'med_comp': 93,
'sup_comp': 22,
'pos_comp': 0})
print(novos_inscritos)
print('-'*32)

# d) Utilizando as duas Series que você tem, calcule o número total de inscritos após a prorrogação do período de inscrições.
print('d) Inscrições totais após prorrogação:')
total_inscricoes = inscritos + novos_inscritos
print(total_inscricoes)
```

## Questão 2

Uma turma de cinco alunos foi submetida a uma avaliação e a nota de cada um dos alunos é apresentada na tabela abaixo:

| Nome do Aluno | Nota |
|---------------|------|
|Wilfred |2|
|Abbie 	|7|
|Harry 	|5.5|
|Julia 	|10|
|Carrie |6.2|

* a) Crie uma Series com os valores da tabela acima.
* b) Retorne o número de alunos que foram aprovados (considerando que a nota mínima para a aprovação é 7).
* c) Qual foi a nota média da turma.
* d) Obtenha o nome dos alunos que tiraram notas acima da média da turma.


```python
notas = pd.Series({'Wilfred': 2,
                   'Abbie': 7,
                   'Harry': 5.5,
                   'Julia': 10,
                   'Carrie': 6.2})

# a)
print('a) Crie uma Series com os valores da tabela')
print(notas)
print('-'*32)

# b)
print('b) Retorne o número de alunos que foram aprovados (considerando que a nota mínima para a aprovação é 7)')
print('Número de alunos aprovados:', notas[notas >= 7].count())
print(notas[notas >= 7])
print('-'*32)

# c)
print('c) Qual foi a nota média da turma')
print(notas.mean())
print('-'*32)

# d)
print('d) Obtenha o nome dos alunos que tiraram notas acima da média da turma')
print('Alunos acima da média da turma:', notas[notas >= notas.mean()].keys())
```

## Questão 3

Crie um objeto DataFrame que contenha os dados da tabela abaixo, os quais se referem ao peso, altura e idade de cinco pessoas:
|Peso (kg)|Altura (cm)|Idade|
|---|---|--|
| 72|180|31|
| 80|170|21|
| 60|175|16|
| 90|174|28|
|100|185|23|

Com base no DataFrame criado, realize o que pedido nos itens abaixo:

* a) Visualize apenas os dados da coluna que contém a altura das pessoas.
* b) Obtenha a idade média das pessoas.
* c) Obtenha os dados das pessoas que estão abaixo da altura média.
* d) Crie um nova coluna no DataFrame que contenha o IMC (Índice de Massa Corporal) de cada pessoa.
* e) Remova a coluna que contém a idade das pessoas.
* f) Salve esses dados do DataFrame em um arquivo csv.


```python
import pandas as pd
import numpy as np

dict_medidas = {'peso': [72, 80, 60, 90, 100],
                'altura': [180, 170, 175, 174, 185],
                'idade': [31, 21, 16, 28, 23]}
df = pd.DataFrame(dict_medidas)
print(df)

print('a) Visualize apenas os dados da coluna que contém a altura das pessoas.')
print(df['altura'])
print('-'*32)

print('b) Obtenha a idade média das pessoas.')
print(df['idade'].mean())
print('-'*32)

print('c) Obtenha os dados das pessoas que estão abaixo da altura média.')
print('altura média:', df['altura'].mean())
print(df[df['altura'] < df['altura'].mean()])
print('-'*32)

print('d) Crie um nova coluna no DataFrame que contenha o IMC (Índice de Massa Corporal) de cada pessoa.')
df['IMC'] = df['peso'] / ((df['altura'] / 100) ** 2)
print(df)
print('-'*32)

print('e) Remova a coluna que contém a idade das pessoas.')
df = df.drop(columns='idade')
print(df)

print('f) Salve esses dados do DataFrame em um arquivo csv.')
df.to_csv('pesos_e_alturas.csv', index = False)
```

### Questão 05

Carregue, em um objeto DataFrame, o mesmo dataset que você salvou na Questão 3. Em seguida,

- a) Adicione uma nova coluna nesse DF chamada Classificação, que contenha a classificação de cada indivíduo, de acordo com o seu IMC. Para isso, consulte a tabela abaixo.

|IMC 	|Classificação 	|Obesidade (grau)|
|-------|---------------|----------------|
|Menor que 18,5| 	Magreza| 	0|
|Entre 18,5 e 24,9| 	Normal| 	0|
|Entre 25,0 e 29,9| 	Sobrepeso| 	I|
|Entre 30,0 e 39,9| 	Obesidade| 	II|
|Maior que 40,0| 	Obesidade Grave| 	III|

- b) Salve em DataFrame no formato xlsx.


```python
import pandas as pd

df = pd.read_csv('pesos_e_alturas.csv')

print('a) Adicione uma nova coluna nesse DF chamada Classificação, que contenha a classificação de cada indivíduo, de acordo com o seu IMC.')

# Definindo as classificações de IMC
def imc(x):
    if x > 40:
        return ('Obesidade Grave', 'III')
    elif x > 30:
        return ('Obesidade', 'II')
    elif x > 25:
        return ('Sobrepeso', 'I')
    elif x > 18.5:
        return ('Normal', '0')
    elif x > 18.5:
        return ('Magreza', '0')
    else:
        return (None, None)

# O retorno é uma tupla. 
# Quando passamos para o apply vira uma Series e então transformamos em uma lista para alimentar as duas colunas

df[['Classificação', 'Obesidade']] = df['IMC'].apply(imc).tolist()

print(df)

print('b) Salve em DataFrame no formato xlsx.')

df.to_excel('imc.xlsx', index=False)
```
